## Idea central

El proyecto consiste en **diseñar un mini-lenguaje cuya validez dependa del contexto global** y construir un sistema que lo verifique mediante un **autómata linealmente acotado (LBA)** o una simulación explícita de dicho modelo.

La intención es pasar de una visión puramente clasificatoria de la teoría de lenguajes a una visión computacional:

- el lenguaje como objeto formal
- el autómata como procedimiento de verificación
- la implementación como hipótesis operativa
- la evaluación como auditoría crítica del propio diseño

# 1. Propósito del proyecto

Este proyecto busca que el estudiante construya una comprensión profunda de los **lenguajes sensibles de contexto (CSL)** y de los **autómatas linealmente acotados (LBA)** no solo como definiciones de libro, sino como modelos efectivos de computación.

El trabajo debe mostrar, de forma integrada:

1. diseño formal de un lenguaje
2. argumentación teórica sobre su clase
3. diseño de una estrategia de reconocimiento
4. implementación computacional en Python
5. análisis experimental
6. autoevaluación rigurosa

---

## Lo que este proyecto NO es

Este proyecto **no** consiste en:

- repetir un ejemplo clásico sin elaboración propia
- dibujar un autómata superficial sin explicar su operación
- hacer únicamente parsing sintáctico
- entregar solo código sin discusión teórica
- asumir que porque “funciona en ejemplos” ya está bien construido
- usar IA para producir respuestas finales sin auditoría conceptual propia

---

## Lo que sí se espera

Se espera un trabajo donde el estudiante construya un sistema que opere sobre una idea genuina de restricción contextual y muestre, con precisión, por qué se requiere memoria acotada linealmente respecto a la entrada.

# 2. Enunciado formal del proyecto

Diseñe un **mini-lenguaje formal** cuya validez dependa de **relaciones globales entre símbolos o segmentos de la entrada**, de modo que su reconocimiento no pueda reducirse razonablemente a un autómata finito ni a un autómata de pila ordinario.

Luego:

1. defina formalmente ese lenguaje
2. diseñe una estrategia de verificación mediante un **autómata linealmente acotado**
3. implemente una simulación en Python
4. analice sus resultados
5. justifique por qué el lenguaje es sensible de contexto o, como mínimo, por qué su propuesta exige memoria contextual de forma no trivial

---

## Condición mínima de interés

El lenguaje propuesto debe incorporar al menos **dos** de estas propiedades:

### A. Dependencia declaración/uso
Un símbolo, variable o recurso debe aparecer o quedar registrado antes de poder usarse.

### B. Balance múltiple
Debe existir una relación de igualdad o correspondencia entre varias regiones de la cadena, por ejemplo una generalización de patrones como $a^n b^n c^n$.

### C. Dependencias cruzadas
La validez de un segmento debe depender de otro segmento distante.

### D. Restricciones semánticas globales
La cadena no puede determinarse como válida observando solo fragmentos locales; hay que considerar la estructura completa.

---

## Nivel esperado

El proyecto debe ser **avanzado, difícil e interesante**.  
No se busca solo corrección técnica, sino una propuesta intelectualmente seria.

# 3. Forma esperada del lenguaje

Su mini-lenguaje puede tomar una forma simbólica, textual o cuasi-programática.

Por ejemplo, puede ser un lenguaje donde:

- ciertos nombres deben ser introducidos y reutilizados con restricciones
- haya varias zonas cuya longitud o estructura deba coincidir
- existan correspondencias entre bloques separados
- haya condiciones globales que obliguen a recorrer y marcar la entrada varias veces
- la validez dependa de relaciones que no puedan resolverse con una sola pila simple

---

## Importante

No basta con escribir una intuición vaga del tipo:

> “mi lenguaje tiene contexto”.

Debe haber una formulación precisa que permita responder:

- ¿cuál es el alfabeto?
- ¿qué cadenas pertenecen al lenguaje?
- ¿qué cadenas no pertenecen?
- ¿cuál es exactamente la restricción contextual?
- ¿cómo la verifica el sistema?

# 4. Componentes obligatorios del trabajo

## 4.1 Definición formal del lenguaje

Debe incluir:

- alfabeto
- forma general de las cadenas
- descripción matemática o formal del lenguaje
- ejemplos válidos
- ejemplos inválidos

---

## 4.2 Diseño conceptual del LBA

Debe incluir:

- descripción de los estados o fases de cómputo
- uso de la cinta
- criterio de aceptación y rechazo
- estrategia de marcaje, comparación o reescritura
- explicación de por qué la memoria usada es lineal respecto a la longitud de la entrada

No es obligatorio presentar una 7-tupla completamente formal si eso empobrece la explicación, pero sí debe quedar claro el diseño computacional del autómata.

---

## 4.3 Implementación en Python

La implementación debe:

- simular el comportamiento de un LBA o una máquina claramente inspirada en ese modelo
- trabajar sobre una cinta acotada linealmente
- procesar entradas válidas e inválidas
- permitir observar, al menos opcionalmente, la evolución del proceso

---

## 4.4 Experimentos

Debe incluir una batería de pruebas que muestre:

- casos válidos
- casos inválidos
- casos límite
- comportamiento esperado del sistema
- observaciones sobre tiempo, pasos o uso de marcas

---

## 4.5 Comparación conceptual

El trabajo debe discutir:

- por qué un DFA no es suficiente
- por qué un PDA ordinario no basta, o por qué sería inadecuado para la propiedad elegida
- por qué un LBA sí resulta adecuado

# 6. Autoevaluación obligatoria

## Este proyecto NO será evaluado por el profesor

La evaluación debe ser realizada por los propios estudiantes.

Esto no significa que la evaluación sea informal o liviana.  
Significa exactamente lo contrario:

> la responsabilidad de demostrar la calidad del trabajo recae en quien lo construye.

La autoevaluación debe ser explícita, argumentada y honesta.

No se aceptará una autoevaluación reducida a una lista de cotejo sin justificación.

# 7. Criterios de autoevaluación

## 7.1 Formulación generativa

Preguntas orientadoras:

- ¿El lenguaje diseñado es realmente interesante?
- ¿Es una simple variación cosmética de un ejemplo clásico?
- ¿La restricción contextual es genuina?
- ¿La formulación obliga a pensar computacionalmente?
- ¿El problema creado tiene suficiente densidad conceptual?

### Qué debe mostrar el estudiante
Debe justificar por qué su propuesta no es trivial y por qué vale la pena estudiarla.

---

## 7.2 Trazabilidad

Preguntas orientadoras:

- ¿Puedo explicar cómo cada parte de la implementación corresponde al diseño del LBA?
- ¿La relación entre teoría, diseño y código es clara?
- ¿Se puede seguir el proceso de verificación paso a paso?
- ¿Mis ejemplos ilustran realmente lo que afirmo?

### Qué debe mostrar el estudiante
Debe haber correspondencia explícita entre:

- definición formal
- estrategia de cómputo
- código
- resultados

---

## 7.3 Consistencia estructural

Preguntas orientadoras:

- ¿Mi lenguaje, mi argumento teórico y mi implementación dicen lo mismo?
- ¿Hay contradicciones entre lo que afirmo y lo que el programa hace?
- ¿Mi sistema acepta cadenas que no debería aceptar?
- ¿Rechaza cadenas que debería aceptar?
- ¿Los ejemplos elegidos son coherentes con la definición?

### Qué debe mostrar el estudiante
Debe identificar con precisión si existe alguna fractura entre el diseño y la ejecución.

---

## 7.4 Auditoría crítica

Preguntas orientadoras:

- ¿Dónde podría estar mal mi argumento?
- ¿Qué supuse sin demostrar?
- ¿Qué casos límite omití?
- ¿Mi justificación de que el lenguaje requiere contexto es fuerte o superficial?
- ¿Qué contraejemplo podría poner en crisis mi diseño?

### Qué debe mostrar el estudiante
Debe examinar críticamente su propio trabajo como si fuera un revisor externo.

# 8. Trabajo recursivo y revisión iterativa obligatoria

Este proyecto **debe trabajarse de forma recursiva e iterativa**.

Eso significa que el estudiante no debe asumir que la primera formulación del lenguaje, del autómata o del código será suficiente.

Se espera explícitamente el siguiente ciclo:

1. formular una primera versión del lenguaje
2. intentar implementarla
3. detectar inconsistencias
4. refinar la definición
5. ajustar el diseño del LBA
6. volver a probar
7. volver a evaluar
8. repetir el proceso hasta cumplir **todo lo indicado en este enunciado**

---

## Insistencia importante

**No basta con cumplir parcialmente.**

Los estudiantes deben trabajar recursivamente **hasta cumplir todo lo indicado**:

- definición formal clara
- diseño computacional convincente
- implementación funcional
- análisis experimental suficiente
- comparación conceptual seria
- autoevaluación argumentada

Si alguno de esos puntos no está sólido, el trabajo **no está terminado**.

---

## Interpretación de “recursivo” en este contexto

Aquí “recursivo” no significa solo una técnica de programación.

Significa que el proyecto debe construirse mediante una revisión constante de sus propios resultados:

- la idea corrige al código
- el código corrige a la idea
- los ejemplos corrigen la definición
- la autoevaluación corrige el proyecto entero

# 9. Uso permitido y recomendado de IA

La IA puede y debe utilizarse como apoyo de pensamiento, pero no como sustituto del juicio del estudiante.

Se recomienda usar IA para:

- discutir si el lenguaje propuesto es trivial o no
- buscar contraejemplos
- refinar definiciones
- cuestionar la estrategia del LBA
- explorar fallos en la implementación
- discutir la calidad de la autoevaluación
- auditar si realmente se cumplen los criterios del proyecto

---

## Muy importante

La IA no debe utilizarse como fuente de cierre automático del proceso.

El hecho de que una IA diga “esto está bien” **no sustituye** la obligación del estudiante de:

- justificar
- contrastar
- probar
- revisar
- corregir

---

## Requisito explícito

Los estudiantes deben, incluso con ayuda de IA, **discutir la evaluación**.

Eso implica que pueden usar IA para preguntarse, por ejemplo:

- ¿mi lenguaje realmente exige contexto?
- ¿mi argumento contra PDA es sólido?
- ¿mi implementación se corresponde con mi explicación?
- ¿mi autoevaluación está siendo complaciente?
- ¿qué huecos conceptuales quedan?

Pero la respuesta final debe estar integrada críticamente en el trabajo propio.

# 10. Lo que debe aparecer en la sección de autoevaluación final

La autoevaluación final debe incluir, de manera explícita:

## A. Juicio sobre la formulación
Qué tan buena, difícil e interesante resultó la propuesta.

## B. Juicio sobre la implementación
Qué tanto el código representa realmente el diseño del LBA.

## C. Juicio sobre la coherencia global
Si el trabajo sostiene lo que afirma.

## D. Fallos encontrados
Errores, limitaciones, dudas abiertas o casos no resueltos.

## E. Discusión crítica con apoyo de IA
Qué objeciones surgieron al discutir el trabajo con IA y cómo fueron resueltas o no resueltas.

---

## Advertencia

Una autoevaluación creíble no es la que se felicita a sí misma, sino la que muestra capacidad de crítica.

# 11. Posibles extensiones de alto nivel

Quien quiera profundizar puede explorar, además:

- no determinismo
- visualización gráfica del LBA
- comparación experimental con intentos de solución mediante PDA
- generadores automáticos de cadenas válidas e inválidas
- análisis del número de pasos o del patrón de marcaje
- discusión más formal sobre la clase del lenguaje

Estas extensiones no reemplazan los requisitos base; solo los elevan.

# 12. Cierre conceptual

Este proyecto no busca solo que el estudiante “reconozca un lenguaje”.

Busca que el estudiante construya una máquina conceptual y computacional que permita responder:

- qué significa que una restricción dependa del contexto
- qué tipo de memoria exige esa dependencia
- cómo se operacionaliza esa verificación
- cómo una implementación revela límites y posibilidades del modelo

En este sentido, el proyecto debe asumirse como un trabajo de construcción rigurosa, revisión constante y evaluación crítica.

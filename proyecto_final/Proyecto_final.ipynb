# Proyecto Final

## Teoría de Autómatas — Aplicaciones Computacionales de Autómatas y Máquinas de Turing

Este curso ha explorado los autómatas no únicamente como objetos matemáticos abstractos, sino como modelos computacionales capaces de representar procesos, sistemas, lenguajes, verificadores, mecanismos de decisión y formas de computación. A lo largo del curso se trabajó con autómatas finitos, expresiones regulares, gramáticas, autómatas de pila, máquinas de Turing, decidibilidad y complejidad computacional, enfatizando siempre la implementación computacional y la construcción de sistemas reales basados en estos modelos.

El proyecto final consiste en el diseño e implementación de una aplicación original basada en alguno de los modelos formales estudiados durante el curso. El objetivo no es únicamente implementar un autómata clásico, sino utilizarlo como mecanismo computacional central dentro de una aplicación novedosa, experimental o conceptualmente interesante.

---

# Objetivo General

Diseñar una aplicación computacional original basada en teoría de autómatas, donde el comportamiento principal del sistema esté gobernado explícitamente por:

* un autómata finito,
* un autómata no determinista,
* un autómata con transiciones ε,
* un autómata de pila,
* una gramática formal,
* una máquina de Turing,
* una máquina de Turing multicinta,
* una máquina no determinista,
* un sistema derivado de complejidad computacional,
* o cualquier combinación de los anteriores.

---

# Restricción Principal

El proyecto debe utilizar explícitamente un modelo formal de computación como parte esencial del sistema.

No se aceptarán proyectos donde el autómata sea únicamente decorativo o una representación superficial.

El modelo formal debe participar activamente en:

* procesamiento,
* reconocimiento,
* decisión,
* simulación,
* generación,
* validación,
* verificación,
* navegación de estados,
* ejecución,
* interpretación,
* o transformación computacional.

# Observación Final
